```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
df['budget'] = df[['TV', 'Radio', 'Newspaper']].sum(axis=1)
df.head()
```

```python
x = df['budget'].values
y = df['Sales'].values
plt.scatter(x, y, color='blue', label='Data Points')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.title('Sales vs Advertising Budget')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
```

```python
Y_pred = model.predict(x.reshape(-1, 1)).reshape(-1, 1)
```

```python
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, Y_pred, color='red', label='Regression Line')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.title('Sales vs Advertising Budget with Regression Line')
```

```python
new_X = np.array([[10000]])
predicted_y = model.predict(new_X)
print(f"Predicted value for X=10000: {predicted_y[0]}")
```
